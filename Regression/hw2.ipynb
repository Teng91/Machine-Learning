# Prob 1

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from sklearn.metrics import r2_score
import warnings
warnings.filterwarnings('ignore')
# read file
f1 = open('02HW1_Xtrain', 'r')
f2 = open('02HW1_Ytrain', 'r')
x = f1.read()
y = f2.read()
f1.close()
f2.close()
# data preprocessing
l1 = list(x.strip().split('\n   '))
X = []
for i in l1:
    X.append(float(i))
l2 = list(y.strip().split('\n  '))
Y = []
for i in l2:
    Y.append(float(i))
# draw scatter plot
plt.scatter(X, Y)
plt.title("scatter plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
# 1st order
model = np.poly1d(np.polyfit(X, Y, 1))
myline = np.linspace(0, 1, 100)
plt.scatter(X, Y)
plt.title("1st order")
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(myline, model(myline))
plt.show()
print("SST:",np.sum((Y - np.mean(X))**2))
print("SSE:",np.sum((Y - model(X))**2))
print("SSR:",np.sum((model(X) - np.mean(X))**2))
print("R2:",1-(np.sum((Y - model(X))**2)/np.sum((Y - np.mean(X))**2)))
# error of 1st order
sns.residplot(X, Y-model(X))
plt.title("error of 1st order")
plt.xlabel("maturities")
plt.ylabel("error values")
plt.show()
# Residual histogram of 1st order
sns.distplot(Y-model(X), bins = 20)
plt.title('Residual histogram of 1st order')           
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.show()
```

```python
model = np.poly1d(np.polyfit(X, Y, 2))
myline = np.linspace(0, 1, 100)
plt.scatter(X, Y)
plt.title("2nd order")
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(myline, model(myline))
plt.show()
print("SST:",np.sum((Y - np.mean(X))**2))
print("SSE:",np.sum((Y - model(X))**2))
print("SSR:",np.sum((model(X) - np.mean(X))**2))
print("R2:",1-(np.sum((Y - model(X))**2)/np.sum((Y - np.mean(X))**2)))
# error of 2nd order
sns.residplot(X, Y-model(X))
plt.title("error of 2nd order")
plt.xlabel("maturities")
plt.ylabel("error values")
plt.show()
# Residual histogram of 2nd order
sns.distplot(Y-model(X), bins = 20)
plt.title('Residual histogram of 2nd order')           
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.show()
```

```python
model = np.poly1d(np.polyfit(X, Y, 3))
myline = np.linspace(0, 1, 100)
plt.scatter(X, Y)
plt.title("3rd order")
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(myline, model(myline))
plt.show()
print("SST:",np.sum((Y - np.mean(X))**2))
print("SSE:",np.sum((Y - model(X))**2))
print("SSR:",np.sum((model(X) - np.mean(X))**2))
print("R2:",1-(np.sum((Y - model(X))**2)/np.sum((Y - np.mean(X))**2)))
# error of 3rd order
sns.residplot(X, Y-model(X))
plt.title("error of 3rd order")
plt.xlabel("maturities")
plt.ylabel("error values")
plt.show()
# Residual histogram of 3rd order
sns.distplot(Y-model(X), bins = 20)
plt.title('Residual histogram of 3rd order')           
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.show()
```

# Prob 3(a)

```python
maturities = np.array([1, 3, 6, 9, 12, 15, 18, 21, 24, 30, 36, 48, 60, 72, 84, 96, 108, 120])
yields = np.array([7.571, 7.931, 7.977, 7.996, 8.126, 8.247, 8.298, 8.304, 8.311, 8.327, 8.369, 8.462, 8.487, 8.492, 8.479, 8.510, 8.507, 8.404])
plt.scatter(maturities, yields)
plt.title('1990.05.31')
plt.xlabel('maturities')
plt.ylabel('yields')
plt.show()
```

從圖中可以看出大致分為三個趨勢：  
1. maturities 介於 1-20 之間，產量呈現大幅度提升  
2. maturities 介於 20-40 之間，產量提升幅度趨緩  
3. maturities = 40之後，產量不再有提升的現象
