```python
import scipy.io
from sklearn import cluster
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
# Problem 1
A = scipy.io.loadmat('11HW1_KmeanData.mat')
print(A['X'].shape)
print(A['Y'].shape)
```

```python
plt.scatter(A['X'][0], A['X'][1], s=20, cmap='viridis');
```

```python
plt.scatter(A['Y'][0], A['Y'][1], s=20, cmap='viridis');
```

```python
X_ = []
for i in range(len(A['X'][0])):
#     print(A['X'][0][i],A['X'][1][i])
    X_.append([A['X'][0][i],A['X'][1][i]])
```

```python
kmeansX = KMeans(n_clusters = 4)
kmeansX.fit(X_)
```

```python
new_X = kmeansX.predict(X_)
new_X
```

```python
plt.subplot(121)
plt.title('Original X')
plt.scatter(np.array(X_).T[0], np.array(X_).T[1], cmap=plt.cm.Set1)
plt.subplot(122)
plt.title('kmeans of X(k=4)')
plt.scatter(np.array(X_).T[0], np.array(X_).T[1], c=new_X, cmap=plt.cm.Set1)
plt.show()
```

```python
import cv2
import matplotlib.image as img
from skimage import io
img_gray = img.imread('11HW2_PatchPanelsGrey.jpeg')
img_color = img.imread('11HW2_PatchPanels.jpg')
plt.subplot(121)
plt.imshow(img_gray)
plt.axis('off')
plt.subplot(122)
plt.imshow(img_color)
plt.axis('off')
plt.show()
print(img_gray.shape)
print(img_color.shape)
```

```python
# compression ratio (Gray)
s_g=665*666*3*8;
# compression ratio (Color)
s_c=480*640*3*8;
```

```python
rows = img_gray.shape[0]
cols = img_gray.shape[1]
image = img_gray.reshape(img_gray.shape[0]*img_gray.shape[1], 3)
kmeans = KMeans(n_clusters = 2, max_iter=200)
kmeans.fit(image)
# 分群結果
clusters = np.asarray(kmeans.cluster_centers_, dtype=np.uint8) 
# 顯示圖片
labels = np.asarray(kmeans.labels_, dtype=np.uint8)  
# 修改圖片尺寸
labels = labels.reshape(rows, cols)
plt.imshow(labels, cmap='gray')
plt.axis('off')
plt.show()
# 計算壓縮比
s_2_g=len(labels)*3*8+666*665*1
c_g_2=s_g/s_2_g
print(c_g_2)
```

```python
rows = img_color.shape[0]
cols = img_color.shape[1]
image = img_color.reshape(img_color.shape[0]*img_color.shape[1], 3)
kmeans = KMeans(n_clusters = 2, max_iter=200)
kmeans.fit(image)
# 分群結果
clusters = np.asarray(kmeans.cluster_centers_, dtype=np.uint8) 
# 顯示圖片
labels = np.asarray(kmeans.labels_, dtype=np.uint8)
# 修改圖片尺寸
labels = labels.reshape(rows, cols)
plt.imshow(labels, cmap='hsv')
plt.axis('off')
plt.show()
# 計算壓縮比
s_2_c=len(labels)*3*8+480*640*1
c_c_2=s_c/s_2_c
print(c_c_2)
```
